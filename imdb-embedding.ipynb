```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("IMDB_Dataset.csv")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
from bs4 import BeautifulSoup
def text_cleaner(text):
    return BeautifulSoup(text).text
    
```

```python
data['review'] = data['review'].apply(text_cleaner)
```

```python
import tensorflow as tf
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer()
```

```python
tokenizer.fit_on_texts(data['review'])
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
tokenizer.word_counts
```

```python
seq_data = tokenizer.texts_to_sequences(data['review'])
```

```python
type(seq_data)
```

```python
seq_data
```

```python
pad_seq = tf.keras.utils.pad_sequences
```

```python
X = pad_seq(seq_data,padding="pre")
```

```python
X.shape
```

```python
y = data['sentiment']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y= le.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from keras.layers import Embedding
```

```python
from keras.layers import Dense
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Conv1D
from keras.layers import Flatten
from keras.layers import Dropout
from keras.regularizers import L2
```

```python
reg = L2(0.000001)
```

```python
model = Sequential()
model.add(Embedding(input_dim= 126507, output_dim=50,input_length=2466,embeddings_regularizer=reg))
model.add(Conv1D(filters=32 , kernel_size=3 , padding= 'same',activation='relu',kernel_regularizer=reg))
model.add(Flatten())
model.add(Dense(64,activation='relu',kernel_regularizer=reg))
model.add(Dropout(0.33))
model.add(Dense(64,activation='relu',kernel_regularizer=reg))
model.add(Dropout(0.33))
model.add(Dense(64,activation='relu',kernel_regularizer=reg))
model.add(Dropout(0.33))
model.add(Dense(64,activation='relu',kernel_regularizer=reg))
model.add(Dense(1,activation='sigmoid',))
```

```python
model.summary()
```

```python
# (3 * 50 +1)*8
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.0001)
los = tf.keras.losses.BinaryCrossentropy()
met = tf.keras.metrics.BinaryAccuracy()
```

```python
model.compile(optimizer = opt , loss = los , metrics = met)
```

```python
model.fit(X_train,y_train,batch_size = 200, epochs=6 ,validation_split= 0.2,verbose=1)
```

```python
model.evaluate(X_test,y_test)
```

```python
model.evaluate(X_train,y_train)
```
